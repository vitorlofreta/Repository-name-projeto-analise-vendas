## Objetivo
Realizar uma análise simples de dados de vendas utilizando Python e visualizar informações importantes como faturamento e produtos mais vendidos.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dados/vendas.csv', sep=';')
```

```python
df.head()
```

```python
df.columns
```

```python
df["data"] = pd.to_datetime(df["data"], dayfirst=True)
```

```python
df["faturamento"] = df["quantidade"] * df["preco"]
```

```python
df.groupby("produto")["quantidade"].sum()
```

```python
df.groupby("produto")["quantidade"].sum().plot(kind="bar")

plt.title("Produtos Mais Vendidos")
plt.xlabel("Produto")
plt.ylabel("Quantidade")

plt.show()
```

```python
faturamento_categoria = df.groupby("categoria")["faturamento"].sum()

print(faturamento_categoria)
```

```python
faturamento_categoria.plot(kind="bar")

plt.title("Faturamento por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Faturamento")

plt.show()
```

```python
faturamento_tempo = df.groupby("data")["faturamento"].sum()

print(faturamento_tempo)
```

```python
faturamento_tempo.plot(kind="line")

plt.title("Faturamento ao Longo do Tempo")
plt.xlabel("Data")
plt.ylabel("Faturamento")

plt.xticks(rotation=45)

plt.show()
```
